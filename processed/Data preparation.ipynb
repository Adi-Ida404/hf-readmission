```python
# Keeping the records only for the heart failure cases
heart_failure_icd9 = set()
heart_failure_icd9 = {'39891','40201','40211','40291','40401','40403','40411','40413','40491','40493','4280','4281','42820',
                      '42821', '42822','42823','42830','42831','42832','42833','42840','42841','42842','42843','4289'}
```

```python
# Step 1: Clean ICD-9 codes (convert all to strings)
crop_diagnosis['icd9_code'] = (
    crop_diagnosis['icd9_code']
    .astype(str)
    .str.replace(r'\.0$', '', regex=True)
)

# Step 2: Aggregate diagnoses per admission
diagnosis_agg = (
    crop_diagnosis
    .groupby(['subject_id', 'hadm_id'])['icd9_code']
    .apply(lambda codes: ','.join(codes.dropna()))
    .reset_index(name='all_diagnosis')
)

# Step 3: Merge aggregated diagnoses into merged_df
merged_df = merged_df.merge(diagnosis_agg, on=['subject_id', 'hadm_id'], how='left')

# Step 4: Fill missing values
merged_df['all_diagnosis'] = merged_df['all_diagnosis'].fillna('No_Diagnoses')

# Step 5: Filter for rows with at least one heart failure ICD-9 code
def has_heart_failure(diagnosis_str):
    codes = diagnosis_str.split(',')
    return any(code in heart_failure_icd9 for code in codes)

# Apply filter
merged_df = merged_df[merged_df['all_diagnosis'].apply(has_heart_failure)]
```

Adding Age column

```python
def calculate_age_robust(dob, admittime):
    """
    Calculate age handling all edge cases:
    - Missing values
    - Invalid dates
    - Date ranges where dob > admittime
    - Ages > 110 years
    - Month/day adjustments
    """
    # Handle missing values
    if pd.isna(dob) or pd.isna(admittime):
        return np.nan
    
    # Ensure datetime (if not already)
    dob = pd.to_datetime(dob, errors='coerce')
    admittime = pd.to_datetime(admittime, errors='coerce')
    
    # Check for conversion failures
    if pd.isna(dob) or pd.isna(admittime):
        return np.nan
    
    # Validate reasonable date ranges
    if (dob.year < 1900) or (dob > admittime):
        return np.nan
    
    # Calculate using date components (most reliable method)
    years = admittime.year - dob.year
    months = admittime.month - dob.month
    days = admittime.day - dob.day
    
    # Adjust if birthday hasn't occurred yet
    if (months < 0) or (months == 0 and days < 0):
        years -= 1
    
    # Return capped age (0-110)
    return max(0, min(years, 110))
```

```python
merged_df['age'] = merged_df.apply(
    lambda row: calculate_age_robust(row['dob'], row['admittime']), 
    axis=1
)
```

Adding column for readmission

```python
# Note: Convert date columns to datetime
merged_df['admittime'] = pd.to_datetime(merged_df['admittime'])
merged_df['dischtime'] = pd.to_datetime(merged_df['dischtime'])
merged_df['dod'] = pd.to_datetime(merged_df['dod'])
```

```python
# Step 2: Define 30-day readmission target
merged_df = merged_df.sort_values(['subject_id', 'dischtime'])
merged_df['next_admittime'] = merged_df.groupby('subject_id')['admittime'].shift(-1)
```

```python
merged_df['days_to_readmit'] = (
    (merged_df['next_admittime'] - merged_df['dischtime']).dt.total_seconds() / 86400 #(60 * 60 * 24)
)

merged_df
```

```python
mask = merged_df['days_to_readmit'].isna()
days_diff = (merged_df['dod'] - merged_df['dischtime']).dt.total_seconds() / 86400

# Assign where value is <= 30
merged_df.loc[mask & (days_diff <= 30) & (days_diff > 0), 'days_to_readmit'] = days_diff - 1
```
